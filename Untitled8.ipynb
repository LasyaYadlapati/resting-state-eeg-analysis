```python
import mne
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import os
```

```python
data_path = "ds004504-download/sub-019/eeg/"
file = os.path.join(data_path, "sub-019_task-eyesclosed_eeg.set")
```

```python
raw = mne.io.read_raw_eeglab(file, preload=True)

raw
```

```python
raw.plot_psd_topo()
```

The first visualization shows the Power Spectral Density (PSD) for each EEG channel, arranged according to the approximate spatial layout of electrodes on the scalp. PSD represents how the signal’s power is distributed across different frequencies, which helps identify dominant neural oscillations in the EEG recording.Across the channels, the spectra generally display a 1/f pattern, where power is highest at low frequencies and gradually decreases as frequency increases. This pattern is typical of neural recordings and reflects the natural spectral structure of brain activity. Several channels also show distinct narrow peaks, which likely correspond to rhythmic oscillations or environmental noise (commonly around power-line frequencies such as 50–60 Hz).The similarity of spectral shapes across channels suggests that the recording captures consistent neural dynamics across multiple electrodes, while small variations between channels may reflect differences in brain region activity or electrode sensitivity.Overall, the PSD topography provides a frequency-domain overview of brain activity, allowing researchers to identify dominant oscillatory bands such as delta, theta, alpha, beta, or gamma rhythms. This step is important for validating signal quality and identifying meaningful neural patterns before further analysis.


```python
data = raw.get_data()
avg_signal = data.mean(axis=0)

plt.figure(figsize=(10,4))
plt.plot(avg_signal)
plt.title("Average EEG Signal Across Channels")
plt.xlabel("Time Points")
plt.ylabel("Amplitude")
plt.show()
```

This plot shows the average EEG signal computed across all channels over time. To generate this visualization, the data from each electrode was first extracted, and then the mean signal across channels was calculated at each time point.The resulting waveform represents the global average brain signal, which highlights overall temporal fluctuations in neural activity while reducing channel-specific noise. The signal oscillates around zero amplitude, which is expected for EEG data after preprocessing. The waveform exhibits rapid fluctuations that reflect the high temporal resolution of neural electrical activity.Although averaging across channels smooths out localized signals, the remaining variability indicates that the brain signal contains ongoing dynamic activity. This type of average signal is often used as a baseline measure of overall neural activity, helping researchers observe general trends or detect large-scale events in the recording.


```python
psd = raw.compute_psd()
freqs = psd.freqs
psd_data = psd.get_data()

alpha_band = (freqs >= 8) & (freqs <= 12)
alpha_power = psd_data[:, alpha_band].mean(axis=1)

plt.bar(range(len(alpha_power)), alpha_power)
plt.title("Alpha Power by Channel")
plt.xlabel("EEG Channel")
plt.ylabel("Power")
plt.show()
```

This graph shows the average power in the alpha frequency band (8–12 Hz) for each EEG channel. Alpha waves are commonly associated with relaxed wakefulness and resting brain activity, particularly in posterior regions of the brain. The bar plot indicates that alpha power varies across channels, suggesting that some electrode locations exhibit stronger alpha activity than others. This variation is expected because neural oscillations are not uniform across the scalp and may reflect differences in activity across brain regions. Analyzing alpha power helps identify where rhythmic brain activity is strongest and is commonly used to study attention, relaxation, and cognitive state.


```python
bands = {
    "Delta": (0,4),
    "Theta": (4,8),
    "Alpha": (8,12),
    "Beta": (12,30),
    "Gamma": (30,45)
}

psd = raw.compute_psd()
freqs = psd.freqs
psd_data = psd.get_data()

band_power = []

for band, (low, high) in bands.items():
    mask = (freqs >= low) & (freqs <= high)
    band_power.append(psd_data[:, mask].mean())

plt.bar(bands.keys(), band_power)
plt.title("Average Brainwave Band Power")
plt.ylabel("Power")
plt.show()
```

This bar chart shows the average power of different EEG frequency bands (Delta, Theta, Alpha, Beta, and Gamma). The delta band has the highest power, indicating that slow brainwave activity dominates the recording. The other bands show much lower power, suggesting weaker activity in faster brainwave frequencies. This helps summarize which brainwave types contribute most to the overall EEG signal.

```python
data = raw.get_data()
sfreq = raw.info["sfreq"]

alpha_band = raw.copy().filter(8,12).get_data()

alpha_power = np.mean(alpha_band**2, axis=0)

time = np.arange(len(alpha_power)) / sfreq

plt.figure(figsize=(10,4))
plt.plot(time, alpha_power)
plt.title("Alpha Power Over Time")
plt.xlabel("Time (seconds)")
plt.ylabel("Alpha Power")
plt.show()
```

Alpha waves are commonly linked to relaxed and resting brain states. The fluctuations in the plot indicate that alpha activity varies throughout the recording, reflecting changes in neural activity over time. This helps show the temporal dynamics of brain oscillations in the alpha band.m

```python
from sklearn.decomposition import PCA

data = raw.get_data().T

pca = PCA(n_components=3)
components = pca.fit_transform(data)

plt.figure(figsize=(8,5))
plt.plot(components[:,0], label="PC1")
plt.plot(components[:,1], label="PC2")
plt.plot(components[:,2], label="PC3")

plt.title("Principal Components of EEG Activity")
plt.xlabel("Time")
plt.ylabel("Component Value")
plt.legend()
plt.show()
```

This graph shows the first three principal components (PC1, PC2, and PC3) obtained from PCA applied to the EEG data. PCA reduces the complex multi-channel signals into a few components that capture the most variance in the data. The fluctuations of these components over time represent the dominant patterns of neural activity across channels. This helps simplify the EEG dataset while preserving the main structure of the brain signal.


```python
data = raw.get_data()
sfreq = raw.info["sfreq"]

signal = data[0]  # first EEG channel
time = np.arange(len(signal)) / sfreq * 1000  # convert to ms

duration_ms = 10000
samples = int(duration_ms * sfreq / 1000)

plt.figure(figsize=(10,6))
plt.plot(time[:samples], signal[:samples])
plt.title("First 10 Seconds of EEG Signal (Channel Fp1)")
plt.xlabel("Time (ms)")
plt.ylabel("Amplitude (µV)")
plt.grid()
plt.show()
```

The signal fluctuates over time, reflecting changes in brain electrical activity. These variations represent normal neural oscillations recorded by the electrode.


```python
segment = signal[:samples]

max_index = np.argmax(segment)

print("Highest amplitude:", segment[max_index])
print("Time:", time[max_index], "ms")
```

```python
filtered = raw.copy().filter(1,40).get_data()[0]

plt.figure(figsize=(10,6))
plt.plot(time[:samples], filtered[:samples])

plt.title("Filtered EEG Signal (1–40 Hz)")
plt.xlabel("Time (ms)")
plt.ylabel("Amplitude (µV)")
plt.grid()
plt.show()
```
